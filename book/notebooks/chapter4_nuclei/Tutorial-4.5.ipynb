# Experimental isoscalar giant monopole resonance (ISGMR)

In this tutorial, you will learn how to extract experimental ISGMR from the toolkit.

---

Import the libraries that will be employed in this tutorial.

```python
# Import numpy
import numpy as np
# Import nucleardatapy package
import nucleardatapy as nuda
```

You can simply print out the properties of the nuda's function that we will use:

```python
# Explore the nucleardatapy module to find the correct attribute
print(dir(nuda.nuc.setupISGMRExp))
```

---

Get the full list of tables:

```python
tables, tables_lower = nuda.nuc.isgmr_exp_tables()
print('tables:',tables)
```

Fix a table

```python
table = '2018-ISGMR-GARG-LATEX'
```

Load the experimental radii for the table:

```python
gmr = nuda.nuc.setupISGMRExp( table = table )
```

print output:

```python
gmr.print_outputs( )
```

plot:

```python
nuda.fig.nuc_setupISGMRExp_fig( None, tables )
```
